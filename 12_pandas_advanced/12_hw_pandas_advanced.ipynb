## Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

если источник traffic_source равен yandex или google, то в source_type ставится organic
для источников paid и email из России - ставим ad
для источников paid и email не из России - ставим other
все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
log = pd.read_csv('visit_log.csv', sep=';')
log.head(10)
```

```python
# последовательно применяем все условия к датафрейму
log.loc[(log.traffic_source == 'yandex') | (log.traffic_source == 'google'), 'source_type'] = 'organic'
log.loc[((log.traffic_source == 'paid') | (log.traffic_source == 'email')) & (log.region == 'Russia'), 'source_type'] = 'ad'
log.loc[((log.traffic_source == 'paid') | (log.traffic_source == 'email')) & (log.region != 'Russia'), 'source_type'] = 'other'
log['source_type'].fillna(log['traffic_source'] , inplace = True) # NaN значения в 'source_type' заменяем на значения из предыдущего столбца
log.head(10)
```

```python
# для проверки группируем данные по значениям из столбца 'source_type'. Выводятся только те, что есть в условии
log.groupby('source_type').count().head()
```

## Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
# читаем файл
import pandas as pd
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
# группируем данные по userId и считаем min/max значения
ratings_grouped = ratings.groupby('userId').agg([min, max])
# находим разницу между max и min
ratings_grouped['diff'] = ratings_grouped['timestamp']['max'] - ratings_grouped['timestamp']['min']
ratings_grouped.head()
```

```python
# создаем датафрейм только с дельтой времени. Индексами будут userId. восстанавливаем их до отдельного столбца
lifetime = ratings_grouped['diff'].reset_index()
lifetime.head()
```

```python
# создаем датафрейм с сумарным количеством оценок

data_q = ratings['userId'].value_counts() # считаем общее кол-во оценок по пользователям

ratings_amount = pd.DataFrame(data_q) # переводим series в датафрейм
ratings_amount = ratings_amount.reset_index() # восстанавливаем индексы
ratings_amount.columns = ['userId', 'quantity'] # переименовываем столбцы

ratings_amount.head()
```

```python
# отбираем только тех пользователей, у которых больше 100 оценок
heavy_users = ratings_amount[ ratings_amount.quantity > 100 ]
heavy_users.head()
```

```python
# теперь нам надо объединить 2 полученных датафрейма: с кол-вом оценок и дельтой времени
# используем левое объединение, за основу берем датафрейм с пользователями, отобранными по большому кол-ву оценок
heavy_users_life = heavy_users.merge(lifetime, on=['userId'], how='left')
heavy_users_life.head(10)
```

```python
# считаем среднюю продолжительность жизни активных пользователей
heavy_users_life['diff'].mean() / 24 / 3600

# получаем, что среднее время активной жизни активных пользователей на сайте - почти 464 дня
```

## Задание 4

Дана статистика услуг перевозок клиентов компании по типам (см файл 12-3_lec_Python_13_join):  
rzd - железнодорожные перевозки  
auto - автомобильные перевозки  
air - воздушные перевозки  
client_base - адреса клиентов.    
Необходимо сформировать две таблицы:  
таблицу с тремя типами выручки для каждого client_id без указания адреса клиента  
аналогичную таблицу по типам выручки с указанием адреса клиента  

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# объединяем датафреймы. Внешнее объединение без потери значений
new_base = rzd.merge(auto, on=['client_id'], how='outer').merge(air, on=['client_id'], how='outer')

"""
Теоретически, для выполнения задачи этого достаточно.
Практически, если мы объединяем данные, значит - это для чего-то нужно, 
т.е. должно давать какую-нибудь новую информацию.
Поэтому можно продолжить: 
"""

# заменяем NaN на 0, т.к. для таких данных логичнее считать суммарные значения, а не средние
new_base.fillna(0, inplace = True)

# добавляем столбец с итогами
new_base['total_revenue'] = new_base.sum(axis=1) - new_base['client_id'] 
# или так: new_base['total_revenue'] = new_base['rzd_revenue'] + new_base['auto_revenue'] + new_base['air_revenue']

# добавляем строку с итогами
new_base.loc['total'] = new_base.sum() 

# сумма client_id смысла не имеет, заменяем ее значение на total
new_base.loc['total','client_id'] = 'total'

# выводим итоговую таблицу
new_base
```

```python
# здесь все то же самое, только объединений больше и без извращений с добавлением итогов. 
# поэтому отсутствующие значения можно заменить на красивые прочерки :) 
new_base_adr = client_base.merge(rzd, on=['client_id'], how='outer').merge(auto, on=['client_id'], how='outer').merge(air, on=['client_id'], how='outer')
new_base_adr.fillna('-', inplace = True)
new_base_adr
```
