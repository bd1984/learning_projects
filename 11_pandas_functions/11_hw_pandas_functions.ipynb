## Задание 1 
Классификация фильмов по рейтингам

```python
# импортируем панду и читаем файл с рейтингами
import pandas as pd
data_r = pd.read_csv('ratings.csv')
data_r.head()
```

```python
# функция, классифицирующая фильмы на основе их рейтинга + результат ее применения в столбце class

def ranking(param):
    if param <=2:
        return "Низкий рейтинг"
    elif param <=4:
        return "Средний рейтинг"
    elif param  > 4:
        return "Высокий рейтинг"

data_r['class'] = data_r['rating'].apply(ranking)
data_r.head(15)    
```

С одной стороны, задание выполнено - функция написана, с другой - она классифицирует не фильмы, а отдельные оценки отдельных пользователей. Поэтому можно извращаться дальше:  

```python
# Найдем средний рейтинг фильмов и запишем данные в отдельный датафрейм + округляем ср.рейтинг
new_data = data_r[['movieId', 'rating']].groupby('movieId').mean().reset_index().round(2)
new_data.head()
```

```python
# читаем файл с фильмами
data_m = pd.read_csv('movies.csv')
data_m.head()
```

```python
# объединяем по userId данные двух датафреймов и получаем новый df, в котором есть названия фильмов и средние рейтинги
data_fin = new_data.merge(data_m, left_on='movieId', right_on='movieId', how='inner')
data_fin.head()
```

```python
# применяем функцию к новому датафрейму
# получаем объединенные данные с названием фльма , средним рейтином и классификацией
data_fin['class'] = data_fin['rating'].apply(ranking)
data_fin.head(10)   
```

```python
# создаем переменную years. Данные сразу переводим в строковые значения

years = []
for i in range (1950, 2011, 1):
    years.append(str(i))
print(years)
```

```python
# пишем функцию, которая вытащит из названия фильма год его создания и запишет в отдельный столбец

def production_year(param):
    for item in years:
        if item in param:
            return item
    else:
        return '1900'

data_fin['year'] = data_fin['title'].apply(production_year)
data_fin.head()
```

```python
# открываем файл
import pandas as pd
data = pd.read_csv('keywords.csv')
data.head()
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] }
geo_data
```

```python
# пишем функцию, которая будет отпределять регион по классификатору geo_data
def reg (keyword): 
    for key, value in geo_data.items():
        for item in value:
            if item in keyword:
                return key
    else:
        return "undefined"
    
data['region'] = data['keyword'].apply(reg)
data.head()
```

```python
# пытаемся понять, все ли нормально посчиталось - группируем данные по регионам - все 3 есть
data.groupby('region').count().head()
```
