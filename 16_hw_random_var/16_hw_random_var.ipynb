## Задание

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

777: 200 руб.  
999: 100 руб.  
555: 50 руб.  
333: 15 руб.  
111: 10 руб.  
х77: 5 руб.  
хх7: 3 руб.  
х00: 2 руб.  
хх0: 1 руб.  

Вопросы:

Выгодна ли игра игроку?
Сколько в среднем приобретает или теряет игрок за одну игру?
**Дополнительное задание повышенной сложности. **

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования



```python
import numpy as np

cost = 1 # Cтоимость игры
prize = 0    # Выигрыш
trials = 525    # Количество попыток
win_sum = 0 # суммарный выигрыш по итогам всех попыток

for i in range(trials):
    random_score = np.random.randint(0, 1000)    # Случайное число от 0 до 999
    if random_score == 777:
        prize = 200
    elif random_score == 999:
        prize = 100
    elif random_score == 555:
        prize = 50
    elif random_score == 333:
        prize = 15
    elif random_score == 111:
        prize = 10
    elif random_score%100 == 77:
        prize = 5
    elif random_score%10 == 7:
        prize = 3
    elif random_score%100 ==0:
        prize = 2
    elif random_score%10 ==0:
        prize = 1
    else:
        prize = 0
    win = prize - cost
    win_sum += win
    
print('Средний выигрыш/проигрыш за 1 игру: ', win_sum/trials)                        
```

При моделировании средняя величина выигрыша/проигрыша может принимать различные значения в зависимости от количества попыток и конкретного набора чисел, выпадающего в случайном распределении. Проигрыш (отрицательные значения) у меня получались гораздо чаще, чем положительные. Последние выпадали пару раз на небольшом числе trials. Так что **игра для игрока, скорее, невыгодна**. По идее, чем больше попыток, тем более точное значение проигрыша можно получить.   
Или его можно рассчитать теоретически:   
Выигрыш = 200р. х вероятность выпадения 777 + 100р. х вер.выпадения 999 + и т.д.   
В ряду от 0 до 999    
777, 999, 555, 333 и 111 встречаются 1 раз - вероятность 0,001   
числа, оканчивающиеся на 77, встречаются 9 раз (точнее, 10, но надо вычесть 777) - вероятность 0,009   
числа, оканчивающиеся на 7, встречаются 90 раз (точнее, 100, но вычитаем 777 и *77) - вероятность 0,09  
числа, оканчивающиеся на 00, встречаются 9 раз - вероятность 0,009  
числа, оканчивающиеся на 0, встречаются 91 раз - вероятность 0,091  
И теперь нужно перемножить выигрыш на вероятность его получить и все сложить. Все остальные события можно не учитывать, т.к. хоть вероятность их и велика, но выигрыш равен нулю, а, значит, и произведение будет равно 0.

```python
av_win = 0.001*200 + 0.001*100 + 0.001*50 + 0.001*15 + 0.001*10 + 0.009*5 + 0.09*3 + 0.009*2 + 0.091*1
total_win = av_win - cost
print('Средний выигрыш, рассчитанный с учетом вероятности выпадения чисел: ', av_win)
print('Средний выигрыш/проигрыш, с учетом того, что игрок платит 1 руб. за игру: ', total_win)
```
