```python
# 1. перечень всех документов
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

# 2.перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

# ф-ция определяет пользователя по номеру документа
def doc_owner(documents, directories):
    doc_number = input('Введите номер документа: ')
    anc_1 = False 
    for doc in documents:
        if doc['number'] == doc_number:
            print(f"Владелец документа {doc['name']}")
            anc_1 = True
    if not anc_1:
        print(f"Документ в базе не найден.")
    return '\n'
# doc_owner(documents, directories)


# ф-ция определяет полку по номеру документа
def doc_shelf(documents, directories):
    doc_number = input('Введите номер документа: ')
    anc_2 = 0
    for k in directories.items():
        for shelf in k:
            for doc in shelf:
                if doc == doc_number:
                    print(f"Документ хранится на полке: {k[0]}")
                    anc_2 = 1
    if anc_2 == 0:
        print('Документ не найден в базе')
    return '\n'
# doc_shelf(documents, directories)


# ф-ция выводит всю информацию по документам
def doc_info(documents, directories):
    for shelf, content in directories.items():
        for doc in documents:
            if doc['number'] in content:
                print(f"№ {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка: {shelf}")
    return '\n'            
# doc_info(documents, directories)
 
                      
# ф-ция добавляет новую полку
def add_shelf(documents, directories):
    shelf_number = input('Введите номер полки: ')
    if shelf_number in directories: 
        print('Такая полка уже существует.')
    else:
        directories[shelf_number] = None
        print('Полка добавлена.')
    # print(f'Текущий перечень полок: {','.join(list(directories.keys()))}')
    print('Текущий перечень полок: ')
    for key in directories:
        print(key, end=', ')
    return '\n'
# add_shelf(documents, directories) 

                      
# ф-ция удаляет полку, если она пустая
def del_shelf(documents, directories):
    shelf_number = input('Введите номер полки: ')
    if shelf_number in directories:
        if len(directories[shelf_number]) !=0:
            print('На полке есть документы. Удалите их перед удалением полки')
        else:
            del (directories[shelf_number])
            print('Полка удалена')
    else: 
        print('Такой полки не существует.')   
    print('Текущий перечень полок:', ','.join(list(directories.keys())))
    return '\n'
# del_shelf(documents, directories)


# основная, все объединяющая, функция
def main(documents, directories): 
    while True:
        user_input = input('Введите команду: ')
        if user_input == 'p':
            print(doc_owner(documents, directories)) # определить владельца по его номеру    
        elif user_input == 's':
            print(doc_shelf(documents, directories))
           # определить полку по номеру            
        elif user_input == 'l':
            print(doc_info(documents, directories))
            # вывести полную инфу по всем док-там            
        elif user_input == 'as':
            print(add_shelf(documents, directories))
            # добавить новую полку           
        elif user_input == 'ds':
            print(del_shelf(documents, directories))
            # удалить существующую полку            
        elif user_input == 'q':
            print('Завершение работы')
            break

main(documents, directories)

```
