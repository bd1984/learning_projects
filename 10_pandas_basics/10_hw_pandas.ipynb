## Задание 1
Скачайте с сайта grouplens.org...movielens/  датасет любого размера.  
Oпределите какому фильму было выставлено больше всего оценок 5.0.  

База - Small: 100,000 ratings and 3,600 tag applications applied to 9,000 movies by 600 users. Last updated 9/2018.

```python
import pandas as pd
data_r = pd.read_csv('ratings.csv')
data_r.head()
# читаем файл с рейтингами
```

```python
data_r.info()
# на всякий случай проверяем, нет ли пропущенных данных. все ОК
```

```python
data_new = data_r[ data_r.rating == 5.0 ]['movieId'].value_counts()
data_new.head(10)
# или data_r.query('rating == 5.0') - фильтруем строки с рейтингом 5.0
# data_r['movieId'].value_counts() - считаем, сколько раз упоминается каждое значение
# на выходе получаем объект Series, для дальнейшей работы его надо превратить в DataFrame
```

```python
# .to_frame() просто использовать нельзя, т.к. будут кривые названия столбцов, поэтому надо:
# сначала обозвать объект датафреймом, потом убрать старые заголовки и назначить новые названия столбцов

df_data_new = pd.DataFrame(data_new)
df_data_new = df_data_new.reset_index()
df_data_new.columns = ['Id', 'quantity']

df_data_new.head()
```

```python
data_m = pd.read_csv('movies.csv')
data_m.head()
# читаем файл с фильмами
```

```python
data_fin = df_data_new.merge(data_m, left_on='Id', right_on='movieId', how='inner')
data_fin.head()
# объединяем по Id данные нового датафрейма с данными из файла с названиями фильмов 
# больше всего оценок 5.0 получил фильм "Побег из Шоушенка"
```

## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
import pandas as pd
data = pd.read_csv('power.csv')
data.head()
# читаем файл с данными
```

```python
baltic_countries = data[ (data['country']=='Lithuania') | (data['country']=='Latvia') | (data['country']=='Estonia')]
baltic_countries.head()
# отбираем Прибалтийские страны
```

```python
filtered_countries = baltic_countries.query('2004 < year < 2011 and 0 <= quantity' )
filtered_countries.head(10)
# добавляем условия: с 2005 по 2010 гг и кол-во больше нуля
```

```python
filtered_countries.sort_values(by=['quantity'], ascending=[True]).head(10)
# фильтруем по кол-ву, чтобы проверить, убрались ли неотрицательные значения. 
```

```python
filtered_countries['year'].unique()
# проверяем, сработал ли отбор по годам
```

```python
final = filtered_countries[ (filtered_countries['category']== 4) | (filtered_countries['category']== 12 ) | (filtered_countries['category']== 21 )]
final.head(10)
# отбираем нужные категории
```

```python
final['category'].unique()
# проверяем отбор категорий
```

```python
final['quantity'].sum()
# суммируем данные по столбцу с количеством. 
# Видимо, это и будет суммарное потребление странами Прибалтики категорий 4,12 и 21 в 2005 - 2010 гг
```
