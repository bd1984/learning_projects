```python
# 1. перечень всех документов
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

# 2.перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

# ф-ция определяет пользователя по номеру документа
def doc_owner(documents, directories):
    doc_number = input('Введите номер документа: ')
    anc = False # здесь и далее вспомогательная переменная, чтобы результаты прохождения цикла не дублировались
    for doc in documents:
        if doc['number'] == doc_number:
            print(f"Владелец документа {doc['name']}")
            anc = True
    if not anc:
        print('Документ в базе не найден.')
    return '\n'
# doc_owner(documents, directories)


# ф-ция определяет полку по номеру документа
def doc_shelf(documents, directories):
    doc_number = input('Введите номер документа: ')
    anc = 0
    for key in directories.items():
        for shelf in key:
            for doc in shelf:
                if doc == doc_number:
                    print(f"Документ хранится на полке: {key[0]}")
                    anc = 1
    if anc == 0:
        print('Документ в базе не найден.')
    return '\n'
# doc_shelf(documents, directories)


# ф-ция выводит всю информацию по документам
def doc_info(documents, directories):
    for shelf, content in directories.items():
        for doc in documents:
            if doc['number'] in content:
                print(f"№ {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка: {shelf}")
    return '\n'            
# doc_info(documents, directories)
 
                      
# ф-ция добавляет новую полку
def add_shelf(documents, directories):
    shelf_number = input('Введите номер полки: ')
    if shelf_number in directories: 
        print('Такая полка уже существует.')
    else:
        directories[shelf_number] = []
        print('Полка добавлена.')
    # print(f'Текущий перечень полок: {','.join(list(directories.keys()))}')
    print('Текущий перечень полок: ')
    for key in directories:
        print(key, end=', ')
    return '\n'
# add_shelf(documents, directories) 

                      
# ф-ция удаляет полку, если она пустая
def del_shelf(documents, directories):
    shelf_number = input('Введите номер полки: ')
    if shelf_number in directories:
        if len(directories[shelf_number]) !=0:
            print('На полке есть документы. Удалите их перед удалением полки')
        else:
            del (directories[shelf_number])
            print('Полка удалена')
    else: 
        print('Такой полки не существует.')   
    print('Текущий перечень полок:', ','.join(list(directories.keys())))
    return '\n'
# del_shelf(documents, directories)

                      
# ф-ция добавляет новый документ 
def add_doc(documents, directories):
    doc_number = input('Введите номер документа: ')
    doc_type = input('Введите тип документа: ')
    doc_holder = input('Введите владельца документа: ')
    doc_folder = input('Введите полку для хранения: ')
        
    if doc_folder in directories: 
        new_doc = {'type':doc_type, 'number':doc_number, 'name':doc_holder}
        documents.append(new_doc)
        directories.setdefault(doc_folder,[]).append(doc_number)
        print('Документ добавлен. Текущий список документов: ')
    else:
        # directories[doc_folder]=[doc_number]
        print('Такой полки не существует. Добавьте полку командой as. Текущий список документов: ')

    for shelf, content in directories.items():
        for doc in documents:
            if doc['number'] in content:
                print(f"№ {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка: {shelf}")
    return '\n'
# add_doc(documents, directories)
                      

# ф-ция удаляет документ 
def del_doc(documents, directories):
    doc_number = input('Введите номер документа: ')
    anc = 0
    # удаление из documents и печать результата
    documents_copy = documents.copy()
    for doc in documents_copy:
        if doc['number'] == doc_number:
            documents.remove(doc)
            print('Документ удален. Текущий список документов: ')
            anc = 1    
    if anc == 0:
        print('Документ не найден в базе. Текущий список документов: ')
    # удаление из directories
    for shelf, content in directories.items():
        if doc_number in content: 
            content.remove(doc_number)  
# печать сводной информации   
    for shelf, content in directories.items():
        for doc in documents:
            if doc['number'] in content:
                print(f"№ {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка: {shelf}")                  
    return ('\n')
# del_doc(documents, directories)                      


# переместить документ с полки на полку
def move_doc(documents, directories):
    doc_number = input('Введите номер документа: ')
    if doc_number in [doc['number'] for doc in documents]:
        doc_folder = input('Введите полку для хранения: ')

        if doc_folder in directories: 
            for docs in directories.values():
                if doc_number in docs:
                    docs.remove(doc_number)         
            directories[doc_folder].append(doc_number)
            print('Документ перемещен. Текущий список документов: ')
            for shelf, content in directories.items():
                for doc in documents:
                    if doc['number'] in content:
                        print(f"№ {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка: {shelf}")
        else: 
            print('такой полки не существует. Текущий перечень полок: ')
            for key in directories:
                print(key, end=', ')                                   
    else:    
        print('Документ не найден в базе. Текущий список документов: ')
        for shelf, content in directories.items():
            for doc in documents:
                if doc['number'] in content:
                    print(f"№ {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка: {shelf}")
    return '\n'
# move_doc(documents, directories)                        
                      
                      
# основная, все объединяющая, функция
def main (documents, directories):
    command_list = {
    'p': doc_owner,  # определить владельца по его номеру    
    's': doc_shelf, # определить полку по номеру            
    'l': doc_info,   # вывести полную инфу по всем док-там            
    'as': add_shelf,  # добавить новую полку           
    'ds': del_shelf, # удалить существующую полку
    'ad': add_doc, # добавить новый документ на существующую полку            
    'd': del_doc, # удалить документ из данных            
    'm': move_doc # переместить документ с полки на полку  
    }
    while True:
        user_input = input('Введите команду: ')
        if user_input in command_list:
            print(command_list[user_input](documents, directories))
        elif user_input == 'q':
            print('Завершение работы.')
            break
        else:
            print('Команды не существует!')

main(documents, directories)                        
```
