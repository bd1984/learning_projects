### Сначала посмотрим качественные данные  

```python
def quality_data_stat(param, quantity):
    print('Модальное значение: ', int(param.value_counts().idxmax()))   # то же самое, что и .mode(), только без вывода типа переменной
    print('Доля, %: ', round((param.value_counts().max() / quantity * 100), 1))

quantity = 300
print('surgery - была ли операция')
quality_data_stat(data['surgery'], quantity)
```

## Базовые статистические показатели для количественных данных

```python
# Для удобства упаковываем все необходимые расчеты в функцию
def statistics(param):

    print('Среднее значение: ', param.round().mean())
    print('Мода: ', param.round().mode()[0])
    print('Медиана: ', param.median())
    print('Максимальное значение: ', param.max())
    print('Минимальное значение: ', param.min())
    print('Размах: ', (param.max() - param.min()))
    print('Среднее квадратическое отклонение: ', param.std())
    print('Дисперсия: ', param.var())
    Q1 = param.quantile(0.25)
    Q3 = param.quantile(0.75)
    print('Q1 - первый квартиль: ', Q1)
    print('Q3 - третий квартиль: ', Q3)
    print('Межквартильный размах: ', (Q3-Q1))

# Подставляем в функцию данные и печатаем с необходимыми пояснениями
print('Ректальная температура', '\n', 'Норма для взрослой лошади - 37.8 С', '\n','-----')
statistics(data['rect_temp'])
```

Какой бы метод подсчета среднего значения мы ни использовали, мы получаем среднюю температуру по больнице. И эта температура соответствует нормам для здоровой взрослой лошади - от 37,8 до 38,4 (если верить некоторым интернет-источникам).  
Так что средние в данном конкретном случае особого смысла не имеют, потому что живые организмы могут оставаться живыми при очень ограниченном диапазоне температуры тела, и критичное отклонение от нормы на пару градусов в "+" или "-" при болезнях случается, видимо, одинаково часто. 

```python
print('Пульс', '\n', 'Норма для взрослой лошади - 30 - 40 ударов/мин', '\n','-----')
statistics(data['pulse'])
```

Средние значения пульса различаются очень сильно, что говорит о большой неоднородности данных. Возможно, в этой ситуации имеет смысл ориентироваться не на их абсолютные значения, а на то, как все данные соотносятся между собой.  
Минимальное значение соответствует нормальному пульсу взрослой лошади в спокойном состоянии.  
Чаще других встречалось значение, чуть превышающее норму (но надо смотреть на долю таких значений, их может быть, например, всего 3 или 5, просто у нас очень большой интервал возможных значений и другие встречались 1-2 раза). 
Медиана превышает норму для здорового животного в 1,5 - 2 раза. И все равно выше нее лежат еще 50 % наблюдений. 
Максимальное значение - 184, что, в среднем, в 5 раз выше нормы. 
Соответственно, у нас очень большие показатели размаха и дисперсии.   
В целом, у больных лошадей сердцебиение может ускоряться до запредельных величин, но не может опускаться ниже нормальных значений здорового животного. 


```python
print('Количество вдохов в минуту', '\n', 'Норма для взрослой лошади - 8 - 10 вдохов/мин', '\n','-----')
statistics(data['resp_rate'])
```

С частотой дыхания точно такая же ситуация, как и с пульсом. Минимальное значение - норма для здорового животного, максимальное значение превышает норму в 10 раз. ### Работа с выбросами для ректальной температуры

```python
# Находим верхнюю и нижнюю границу  
q1 = data['rect_temp'].quantile(0.25)
q3 = data['rect_temp'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
print('Нижняя граница: ', lower_bound)
print('Верхняя граница: ', upper_bound)

# Находим датафрейм без выбросов
# Заодно выпадают и строки с пустыми значениями (они не влияют на расчет средних). Помним о них но пока игнорируем
remove_outliers_rt = data[data['rect_temp'].between(lower_bound, upper_bound, inclusive=True)].sort_values('rect_temp')
remove_outliers_rt
```

```python
# Теперь посмотрим, насколько отличаются расчет статистик для общей базы и данных без выбросов
print('Среднее без выбросов: ', remove_outliers_rt['rect_temp'].mean())
print('Медиана без выбросов: ', remove_outliers_rt['rect_temp'].median())
print('Разница между средним и средним без выбросов: ', data['rect_temp'].mean() - remove_outliers_rt['rect_temp'].mean())
print('Разница между медианой и медианой без выбросов: ', data['rect_temp'].median() - remove_outliers_rt['rect_temp'].median())
```

Выбросы практически не влияют на средние значения. Поэтому их можно оставить. Тем более, если более подробно изучить эти самые выбросы (см. ниже), то окажется, что их всего 14, из них половина выпадает за нижнюю границу и половина - за верхнюю.

```python
# Ищем строки с выбросами (сливаем датафреймы, удаляем дубликаты и строки с пустыми значениями по столбцам)
rect_temp_outl = pd.concat([data, remove_outliers_rt]).drop_duplicates(keep=False)
rect_temp_outl = rect_temp_outl[rect_temp_outl['rect_temp'].notna()]
rect_temp_outl
```
